```python
#catorgarizing land price 
def landprice(price):
    if price>= 2000000 and price< 4096500:
        return 1
    elif  price>= 4096500 and price< 6700000:
        return 2
    elif  price>= 6700000 and price< 11200000:
        return 3
    else:
        return 4
    
    
Df_dataset["Price"]=Df_dataset["Price"].apply(landprice)

Df_dataset.head()
```

```python
#dropping low predictive coloumns
Df_dataset.drop(['Location','School','JoggingTrack','RainWaterHarvesting','IndoorGames','Intercom','SportsFacility','ATM','ClubHouse','CarParking','MultipurposeRoom','WashingMachine','Gasconnection','Wifi',"Children'splayarea",'LiftAvailable'], axis=1, inplace=True)
```

```python
Df_dataset
```

```python
#method 1 - remove missing values by using created method.
#Remove columns which hv more than 30% of missing values
col_num=0
TotalObjects = Df_dataset.shape[0]
print("Column\t\t\t\t\t Null Values%")
for x in Df_dataset:
    nullCount = Df_dataset[x].isnull().sum();
    nullPercent = nullCount*100/(TotalObjects)
    if nullCount > 0 and nullPercent > 30:
        col_num*col_num+1
        Df_dataset.drop(x, axis=1,inplace=True)
        print(str(x)+"\t\t\t\t\t"+ str(nullPercent))
print("A total of "+str(col_num)+" deleted !")
    
```

```python
#check coloumns with more null values
Df_dataset.info()
```

```python
#method 2 - Replace null values
Df_dataset['SwimmingPool'].fillna(5,inplace = True)
Df_dataset['ShoppingMall'].fillna(5,inplace = True)
#Df_dataset['School'].fillna(5,inplace = True)
Df_dataset['24X7Security'].fillna(5,inplace = True)
Df_dataset['PowerBackup'].fillna(5,inplace = True)
Df_dataset['Hospital'].fillna(5,inplace = True)
Df_dataset['AC'].fillna(5,inplace = True)
Df_dataset['Furnished'].fillna(5,inplace = True)
```

```python
Df_dataset.info()
```

```python
#check if we have null values 
Df_dataset.isnull().sum()
```

```python
#converting all colomns to int 
Df_dataset= Df_dataset.astype({'SwimmingPool':'int64','ShoppingMall':'int64','24X7Security':'int64','PowerBackup':'int64','Hospital':'int64','AC':'int64','Furnished':'int64'})

Df_dataset.dtypes
```

```python
#dealing with catergorical data
def getNumber(str):
    if str=="Banglore":
        return 1
    elif str=="Chennai":
        return 2
    elif str=="Delhi":
        return 3
    elif str=="Hyderabad":
        return 4
    elif str=="Kolkata":
        return 5
    elif str=="Mumbai":
        return 6
    
        
Df_dataset["City"]=Df_dataset["City"].apply(getNumber)
Df_dataset['City']
```

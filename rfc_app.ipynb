```python

import numpy as np
import pickle
import pandas as pd
import streamlit as st 
from PIL import Image
import base64


pickle_in = open("rfc_classifier.pkl","rb")
classifier=pickle.load(pickle_in)



def predict_price(Area,City,Noof_Bedrooms,Resale,SwimmingPool,ShoppingMall ,Security ,PowerBackup ,Hospital ,AC ,Furnished):
   
    prediction=classifier.predict([[Area,City,Noof_Bedrooms,Resale,SwimmingPool,ShoppingMall ,Security ,PowerBackup ,Hospital ,AC ,Furnished]])
    print(prediction)
    return prediction



def main():
    
    st.title("India Property Prices")
    html_temp = """
    

    
Land Price Predictor

    

   
    """
    
    st.markdown(html_temp,unsafe_allow_html=True)
    
    page_bg_img = '''
    
    body {
    background-image:("index.jpg");
    background-size: cover;
    }
    
    '''

    st.markdown(page_bg_img, unsafe_allow_html=True)
    
    
    Area = st.number_input("Area")
    location_select = st.selectbox('Select the location',['Banglore', 'Kolkata', 'Chennai', 'Mumbai',
        'Hyderabad', 'Delhi'])

    locations = {'Banglore': 1,
    'Kolkata': 5,
    'Chennai': 2,
    'Mumbai': 6,
    'Hyderabad': 4,
    'Delhi': 3}


    City = locations[location_select]

    resale_select = st.selectbox('New/Resale',['Resale', 'New',])
    resales = {'Resale': 1,
    'New':0}
    Resale = resales[resale_select]

    Noof_Bedrooms = st.slider('bedroom',0,10)

    sp_select = st.selectbox('Option:Swimming Pool',['Includes Pool', 'No Pool',])
    sp= {'Includes Pool': 1,
    'No Pool':0}
    SwimmingPool = sp[sp_select]


    sm_select = st.selectbox('Option:Shopping Mall',['Shopping Mall in close vicinity', 'No Shopping Mall',])
    sm= {'Shopping Mall in close vicinity': 1,
    'No Shopping Mall':0}


    ShoppingMall = sm[sm_select]
    
    sec_select = st.selectbox('Option:Security',['24x7 Security', 'No Security',])
    sec= {'24x7 Security': 1,
    'No Security':0}
    Security = sec[sec_select]
    
    pb_select = st.selectbox('Option:Power Backup',['Includes Backup', 'No Backup',])
    pb= {'Includes Backup': 1,
    'No Backup':0}
    PowerBackup = pb[pb_select]

    hos_select = st.selectbox('Option:Hospital',['Hospital in Close vicinity', 'No Hospital',])
    hos= {'Hospital in Close vicinity': 1,
    'No Hospital':0}
    Hospital = hos[hos_select]

    ac_select = st.selectbox('Option:Air Conditioning',['Air Conditioned', 'No Air Condition',])
    ac= {'Air Conditioned': 1,
    'No Air Condition':0}
    AC = ac[ac_select]

    furnished_select = st.selectbox('Furnished/Not Furnished',['Furnished', 'Not Furnished',])
    f= {'Furnished': 1,
    'Not Furnished':0}
    Furnished = f[furnished_select]
    
    result=""
    
    normal_html="""  
      
       
 The house is in the price range of 2,000,000 INR - 4,096,500 INR

       

    """
    standard_html="""  
      
       
 The house is in the price range of 4,096,500 INR - 6,700,000 INR

       

    """
    semi_luxury_html="""  
      
        
 The house is in the price range of 6,700,000 INR -11,200,000 INR

       

    """
    luxury_html="""  
      
       
 The house is in the price range of 11,200,000 INR - 40,000,000 INR 

       

    """
    
    
    
    if st.button("Predict"):
        result=predict_price(Area,City,Noof_Bedrooms,Resale,SwimmingPool,ShoppingMall ,Security ,PowerBackup ,Hospital ,AC ,Furnished)
        st.success('The output is {}'.format(result))
        
        if result == 1:
            st.markdown(normal_html,unsafe_allow_html=True)
        elif result == 2:
            st.markdown(standard_html,unsafe_allow_html=True)
        elif result == 3:
            st.markdown(semi_luxury_html,unsafe_allow_html=True)
        else:
            st.markdown(luxury_html,unsafe_allow_html=True)

        

if __name__=='__main__':
    main()
   
 
```
